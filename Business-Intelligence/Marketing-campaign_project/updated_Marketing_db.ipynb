**DESCRIPTION DES VARIABLES**

**add.id**
Identifiant unique de la publicité :
Chaque ligne du dataset représente une publicité ou une campagne spécifique, identifiée de manière unique.

**date.add**
Date de lancement de la publicité :
Permet d’analyser les performances dans le temps (effet de saisonnalité, jour de la semaine, etc.).

**Incl.image**
Inclusion d'une image dans la publicité (1 = oui, 0 = non) :
Permet de tester si les publicités avec images performent mieux que celles sans.

**bounce.rate**
Taux de rebond :
Pourcentage de visiteurs qui quittent le site sans interagir davantage. Un taux élevé signifie souvent que la pub ou la page d’atterrissage ne capte pas l’intérêt.

**rev.per.vis (Revenue per visitor)**
Revenu moyen généré par visiteur :
Mesure le retour sur investissement par visiteur, très utile pour juger de l'efficacité de la campagne.

**conv2customer (Conversion en client)**
Taux de conversion des visiteurs en clients :
Indique la proportion de visiteurs qui ont effectué un achat.

**conv2lead (Conversion en lead)**
Taux de conversion en prospect (lead) :
Un lead peut être une inscription à une newsletter, un remplissage de formulaire, etc. Utile dans des campagnes B2B.

**Incl.disc**
Présence d'une réduction ou d'une promotion (1 = oui, 0 = non) :
Indique si l’annonce propose une remise ou offre spéciale.

**uses.cookies**
Utilisation de cookies de suivi (1 = oui, 0 = non) :
Indique si des cookies ont été utilisés pour le retargeting ou l’analyse comportementale.

**CPC (Cost Per Click)**
Coût par clic :
Montant moyen dépensé par clic sur la publicité. Permet de suivre le coût d’acquisition.

**n.reach2**
Nombre estimé de personnes atteintes par la campagne :
C’est la portée (reach) de la publicité — important pour analyser la visibilité de la campagne.

**Country**
Pays dans lequel la campagne a été diffusée :
Permet des comparaisons géographiques (performances par pays).

**platf (Plateforme)**
Plateforme utilisée pour diffuser la publicité (ex. Facebook, Instagram, Google Ads…) :
Cela peut influencer la performance (par ex. TikTok attire un public plus jeune).



```python
# Conversion de la colonne 'date.add' au format DateTime
marketing['date.add']= pd.to_datetime(marketing['date.add'], errors= 'coerce')

# Ajout d'une colonne 'year'
marketing['year']= marketing['date.add'].dt.year

# Ajout d'une colonne 'month'
marketing['month']= marketing['date.add'].dt.month

```

```python
# Renommer des colonnes
marketing = marketing.rename(columns={
    'rev.per.vis': 'revenue per visitor',
    'bounce.rate': 'bounce rate',
    'conv2customer': 'customer conversion',
    'conv2lead': 'lead conversion',
    'Incl.image': 'included image',
    'Incl.disc': 'included discount',
    'uses.cookies': 'used cookies',
    'n.reach2': 'nb reach',
    'platf': 'Plateforme',
})
```

```python
# Convertir la colonne 'included image' au format booléen
marketing['included image'] = marketing['included image'].astype('bool')
print(marketing['included image'].dtypes)

# Convertir la colonne 'included discount' au format booléen
marketing['included discount'] = marketing['included discount'].astype('bool')
print(marketing['included image'].dtypes)

# Convertir la colonne 'used cookies' au format booléen
marketing['used cookies'] = marketing['used cookies'].astype('bool')
print(marketing['included image'].dtypes)


# Remplace O et 1 par no image et image dans la colonne 'Included image'
marketing['included image']= marketing['included image'].replace({False: 'no image', True: 'image'})

# Remplace O et 1 par no discount et discount dans la colonne 'included discount'
marketing['included discount']= marketing['included discount'].replace({False: 'no discount', True: 'discount'})

# Remplace O et 1 par no  et yes dans la colonne 'used cookies'
marketing['used cookies']= marketing['used cookies'].replace({False: 'no', True: 'yes'})
```

```python
# Faire un arrondi des colonnes 'bounce rate', 'revenue per visitor', 'customer conversion', 'lead conversion'
marketing['bounce rate'] = marketing['bounce rate'].round(2)
marketing['revenue per visitor'] = marketing['revenue per visitor'].round(2)
marketing['customer conversion'] = marketing['customer conversion'].round(2)
marketing['lead conversion'] = marketing['lead conversion'].round(2)
marketing['CPC'] = marketing['CPC'].round(2)
```

```python
##Création de colonnes utiles

# Conversion totale (customer + lead)
marketing['conversion totale'] = marketing['customer conversion'] + marketing['lead conversion']

# Taux de conversion (customer / lead)
marketing['ratio customer/lead'] = marketing['customer conversion'] / marketing['lead conversion'] 

# Revenu total estimé par campagne
marketing['predicted revenue'] = marketing['revenue per visitor'] * marketing['nb reach']

# Coût estimé total par campagne
marketing['predicted cost'] = marketing['CPC'] * marketing['nb reach']

# Rentabilité estimée (Différence entre revenu estimé et coût)
marketing['predicted profit'] = marketing['predicted revenue'] - marketing['predicted cost']

# ROI estimée (Taux, mesure relative de performance. Dit combien on gagne par euro investi.)
marketing['predicted ROI'] = (marketing['predicted revenue'] - marketing['predicted cost']) / marketing['predicted cost']

# Revenu par clic
marketing['revenue per clic'] = marketing['revenue per visitor'] / marketing['CPC']

# Impact visuel / promo
marketing['campaign type'] = marketing['included image'] + '_' + marketing['included discount']
```

```python
# Faire un arrondi des colonnes 'bounce rate', 'revenue per visitor', 'customer conversion', 'lead conversion'
marketing['conversion totale'] = marketing['conversion totale'].round(2)
marketing['ratio customer/lead'] = marketing['ratio customer/lead'].round(2)
marketing['predicted revenue'] = marketing['predicted revenue'].round(2)
marketing['predicted cost'] = marketing['predicted cost'].round(2)
marketing['predicted profit'] = marketing['predicted profit'].round(2)
marketing['predicted ROI'] = marketing['predicted ROI'].round(2).astype(str) + '%'
marketing['revenue per clic'] = marketing['revenue per clic'].round(2)
```
